```python
anime_most_popular = df_anime.sort_values('Members', ascending=False)[:20]
sns.barplot(x='Members', y='Name', data=anime_most_popular, palette='dark')
plt.title("Most popular anime")
plt.show()
```

```python
for type in df['type'].unique():

    anime_highest_rated = df[df['type'] == type].sort_values('score', ascending=False).head(5)
    plt.figure(figsize=(10,5))
    sns.barplot(x='score', y='title', data=anime_highest_rated, palette='plasma')
    plt.title(f"Highest rated anime of type {type}")
    plt.tight_layout()
    plt.show()
```

```python
# Create a scatter plot
fig = px.scatter(df_anime, x='Score', y='Members', 
                 labels={'Score':'Overall Score', 'Members':'Number of Scores'}, 
                 title='Anime Score vs. Number of Scores')

fig.show()
```
